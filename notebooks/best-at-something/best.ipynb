```python
from scipy.stats import binom
import numpy as np
import matplotlib.pyplot as plt
import plots
```

```python
p = 1 - 0.99
n = 500
population = 7_000_000_000
```

```python
1 - (1-p) ** n
```

```python
x = range(0, 10, 1)
cdf = binom(n, p).cdf(x)

plt.plot(x, 1 - cdf, '--o')
plt.xlabel("$M$", size=16)
plt.ylabel("$1 - \sum_{i=0}^M B(i;N,p)$", size=16)

plt.savefig("../../docs/best-at-something/distribution.svg")
```
