```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 1000)
y1 = 1/2 * np.exp(-np.abs(x))
y2 = 1 / np.sqrt(2 * np.pi) * np.exp(-1/2 * np.abs(x)**2)
```

```python
plt.figure(figsize=(5, 5))

plt.plot(x, y1, 'k-', label=r"Laplace Prior")
plt.plot(x, y2, 'k--', label="Gauss Prior")

plt.xlabel(r"$ \beta $ ", size=12)
plt.ylabel("density", size=12)

plt.xticks(size=12)
plt.yticks(size=12)

plt.legend()

plt.savefig("laplace-vs-gauss.svg")
```
